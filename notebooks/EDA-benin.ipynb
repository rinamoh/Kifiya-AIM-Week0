```python
#Time Series Analysis 
#for this analysis since we will be using the column timestamp frequently i will change it to datetime format for easy manipulation of data
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
#and then set it as an index in our table to make it easier to plot
data.set_index('Timestamp', inplace=True)
```

```python
#plot GHI
plt.figure(figsize=(14, 6))
plt.plot(data.index, data['GHI'], label='GHI (W/m²)', color='orange')
plt.title('Global Horizontal Irradiance (GHI) Over Time')
plt.xlabel('Date')
plt.ylabel('GHI (W/m²)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```

```python
 # to evaluate the impact of cleaning on moda and modb data we first need to filter out the cleaned data
cleaned_data = data[data['Cleaning'] == 1]
```

```python
#effect of cleaned data on moda
plt.figure(figsize=(14, 6))
plt.plot(data.index, data['ModA'], label='ModA Readings', color='purple', alpha=0.5)
plt.scatter(cleaned_data.index, cleaned_data['ModA'], color='black', label='Cleaning Events', zorder=5)
plt.title('ModA Sensor Readings with Cleaning Events')
plt.xlabel('Date')
plt.ylabel('ModA (W/m²)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.plot(data.index, data['ModB'], label='ModB Readings', color='cyan', alpha=0.5)
plt.scatter(cleaned_data.index, cleaned_data['ModB'], color='orange', label='Cleaning Events', zorder=5)
plt.title('ModB Sensor Readings with Cleaning Events')
plt.xlabel('Date')
plt.ylabel('ModB (W/m²)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```

```python
#analysing the effects of cleaning by creating our own columns
data['Before Cleaning'] = data['Cleaning'].shift(1).fillna(0).astype(int)
average_readings = data.groupby('Cleaning').agg({
    'ModA': ['mean'],
    'ModB': ['mean']
}).reset_index()
average_readings.columns = ['Cleaning', 'Avg ModA', 'Avg ModB']
print(average_readings)
```

```python
#let us do a graph to show the impact of cleaning to visualize it
plt.figure(figsize=(10, 6))
x = ['No Cleaning', 'Cleaning']
plt.bar(x, average_readings['Avg ModA'], label='Avg ModA (W/m²)', alpha=0.7)
plt.bar(x, average_readings['Avg ModB'], label='Avg ModB (W/m²)', alpha=0.7, bottom=average_readings['Avg ModA'])
plt.title('Impact of Cleaning on Sensor Readings')
plt.ylabel('Average Sensor Readings (W/m²)')
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
#Correlation Analysis on GHI,DNI,DHI AND TModA plus TModB
correlation_data = data[['GHI', 'DNI', 'DHI', 'TModA', 'TModB']]
```

```python
#calculate the correlation matrix 
correlation_matrix = correlation_data.corr()
print(correlation_matrix)

#as you can see from the following table we have the strongest relation ship between GHI and both of the tempratures indicated by approximatly 
#0.9 for TModA and 0.8 for TModB
```

```python
#we can also plot the correlation using heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Matrix of Solar Radiation Components and Temperature Measures')
plt.show()
```
